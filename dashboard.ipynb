```python
import pandas as pd

# Load all datasets (adjust path if needed)
sales_df = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\datasets\SalesFINAL12312016.csv')
purch_df = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\datasets\PurchasesFINAL12312016.csv')
beg_inv_df = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\datasets\BegInvFINAL12312016.csv')
end_inv_df = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\datasets\EndInvFINAL12312016.csv')

# Aggregate metrics
sales_agg = sales_df.groupby('InventoryId', as_index=False)['SalesQuantity'].sum().rename(columns={'SalesQuantity': 'TotalSales'})
purch_agg = purch_df.groupby('InventoryId', as_index=False)['Quantity'].sum().rename(columns={'Quantity': 'TotalPurchases'})
beg_inv_agg = beg_inv_df.groupby('InventoryId', as_index=False)['onHand'].sum().rename(columns={'onHand': 'BeginInventory'})
end_inv_agg = end_inv_df.groupby('InventoryId', as_index=False)['onHand'].sum().rename(columns={'onHand': 'EndInventory'})

# Merge all together
df = sales_agg.merge(purch_agg, on='InventoryId', how='inner')
df = df.merge(beg_inv_agg, on='InventoryId', how='inner')
df = df.merge(end_inv_agg, on='InventoryId', how='inner')

# Calculate demand and reorder
df['NetDemand'] = df['TotalSales'] - df['TotalPurchases']
df['RecommendedReorder'] = df['NetDemand'].clip(lower=0)

# Add store/product info from sales_df (optional: first match only)
product_info = sales_df[['InventoryId', 'Store', 'Brand', 'Description', 'Size']].drop_duplicates(subset='InventoryId')
df = df.merge(product_info, on='InventoryId', how='left')

# Reorder columns for readability
df = df[['InventoryId', 'Store', 'Brand', 'Description', 'Size',
         'TotalSales', 'TotalPurchases', 'BeginInventory',
         'EndInventory', 'NetDemand', 'RecommendedReorder']]

# Preview enriched data
print("‚úÖ Final enriched dataset preview:\n")
print(df.head(10))
```

```python
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# -------- File Paths --------
sales_path = r'C:\Users\user\Desktop\wine_inventory\datasets\SalesFINAL12312016.csv'
begin_path = r'C:\Users\user\Desktop\wine_inventory\datasets\BegInvFINAL12312016.csv'
end_path = r'C:\Users\user\Desktop\wine_inventory\datasets\EndInvFINAL12312016.csv'

# -------- Load & Prepare --------
sales_df = pd.read_csv(sales_path)
begin_df = pd.read_csv(begin_path)
end_df = pd.read_csv(end_path)

# Convert dates
sales_df['SalesDate'] = pd.to_datetime(sales_df['SalesDate'])
begin_df['startDate'] = pd.to_datetime(begin_df['startDate'])
end_df['endDate'] = pd.to_datetime(end_df['endDate'])

# Dropdowns
store_dropdown = widgets.Dropdown(
    options=['All'] + sorted(sales_df['Store'].dropna().unique()),
    value='All',
    description='Store:'
)

brand_dropdown = widgets.Dropdown(
    options=['All'] + sorted(sales_df['Brand'].dropna().unique()),
    value='All',
    description='Brand:'
)

# Filter and plot function
def plot_combined_trends(store, brand):
    # Filter sales
    temp_sales = sales_df.copy()
    if store != 'All':
        temp_sales = temp_sales[temp_sales['Store'] == store]
    if brand != 'All':
        temp_sales = temp_sales[temp_sales['Brand'] == brand]
    
    sales_trend = (
        temp_sales.groupby(pd.Grouper(key='SalesDate', freq='M'))['SalesQuantity']
        .sum()
        .reset_index()
        .rename(columns={'SalesDate': 'Month', 'SalesQuantity': 'TotalSales'})
    )

    # Filter begin
    temp_begin = begin_df.copy()
    if store != 'All':
        temp_begin = temp_begin[temp_begin['Store'] == store]
    if brand != 'All':
        temp_begin = temp_begin[temp_begin['Brand'] == brand]
    begin_trend = (
        temp_begin.groupby(pd.Grouper(key='startDate', freq='M'))['onHand']
        .sum()
        .reset_index()
        .rename(columns={'startDate': 'Month', 'onHand': 'BeginInventory'})
    )

    # Filter end
    temp_end = end_df.copy()
    if store != 'All':
        temp_end = temp_end[temp_end['Store'] == store]
    if brand != 'All':
        temp_end = temp_end[temp_end['Brand'] == brand]
    end_trend = (
        temp_end.groupby(pd.Grouper(key='endDate', freq='M'))['onHand']
        .sum()
        .reset_index()
        .rename(columns={'endDate': 'Month', 'onHand': 'EndInventory'})
    )

    # Merge all
    df = pd.merge(sales_trend, begin_trend, on='Month', how='outer')
    df = pd.merge(df, end_trend, on='Month', how='outer')
    df = df.sort_values('Month')

    # Plot
    fig = px.line(
        df, x='Month', y=['TotalSales', 'BeginInventory', 'EndInventory'],
        title=f'Monthly Trends - Store: {store}, Brand: {brand}',
        labels={'value': 'Units', 'Month': 'Month', 'variable': 'Metric'}
    )
    fig.update_traces(mode='lines+markers')
    fig.update_layout(height=500)
    fig.show()

# Display
widgets.interact(plot_combined_trends, store=store_dropdown, brand=brand_dropdown)
```

```python
# üì¶ Step 3 - Pricing Optimization: Sales vs. Price Sensitivity
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display

# üëâ Load your original sales file (change this to your actual path)
sales_data = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\datasets\SalesFINAL12312016.csv')

# üßπ Basic cleaning: remove zero or negative prices
sales_data = sales_data[sales_data['SalesPrice'] > 0]
sales_data = sales_data[sales_data['SalesQuantity'] > 0]

# üß† Widget filters
store_dropdown = widgets.Dropdown(
    options=['All'] + sorted(sales_data['Store'].unique()),
    value='All',
    description='Store:'
)

brand_dropdown = widgets.Dropdown(
    options=['All'] + sorted(sales_data['Brand'].unique()),
    value='All',
    description='Brand:'
)

# üìâ Scatter plot function
def show_price_sensitivity(store, brand):
    df = sales_data.copy()
    if store != 'All':
        df = df[df['Store'] == store]
    if brand != 'All':
        df = df[df['Brand'] == brand]

    fig = px.scatter(
        df,
        x='SalesPrice',
        y='SalesQuantity',
        color='Description',
        hover_data=['InventoryId', 'SalesDollars'],
        title=f'üí∞ Price Sensitivity (Filtered by Store: {store}, Brand: {brand})',
        opacity=0.6
    )
    fig.update_layout(height=600)
    fig.show()

# üéõÔ∏è Display interactive widget
widgets.interact(show_price_sensitivity, store=store_dropdown, brand=brand_dropdown)
```

```python
# üìä Part 2 ‚Äì Pricing Bands Summary per Product
# Group by InventoryId to see price distribution across products

pricing_summary = (
    sales_data.groupby(['InventoryId', 'Brand', 'Description'])
    .agg(
        MinPrice=('SalesPrice', 'min'),
        MedianPrice=('SalesPrice', 'median'),
        MaxPrice=('SalesPrice', 'max'),
        AvgSales=('SalesQuantity', 'mean'),
        TotalUnitsSold=('SalesQuantity', 'sum'),
        TotalSales=('SalesDollars', 'sum')
    )
    .reset_index()
    .sort_values(by='TotalUnitsSold', ascending=False)
)

# Display top 20 products by units sold
print("üìå Suggested Pricing Bands for Top 20 Products:\n")
display(pricing_summary.head(20))
```

```python
# üß† Add Price Elasticity Bucket based on sales & pricing
def categorize_elasticity(row):
    if row['MedianPrice'] >= 20 and row['TotalUnitsSold'] < 10:
        return 'Overpriced'
    elif row['MedianPrice'] <= 5 and row['TotalUnitsSold'] > 50:
        return 'Undervalued'
    else:
        return 'Fair Priced'

# Apply categorization
pricing_summary['ElasticityCategory'] = pricing_summary.apply(categorize_elasticity, axis=1)

# Show distribution
category_counts = pricing_summary['ElasticityCategory'].value_counts()
print("üìä Product Pricing Categories:")
display(category_counts)

# Optional: Visualize
import plotly.express as px

fig = px.scatter(
    pricing_summary, 
    x='MedianPrice', 
    y='TotalUnitsSold',
    color='ElasticityCategory',
    hover_data=['Brand', 'Description'],
    title='üßÆ Price Elasticity Analysis: Units Sold vs. Price'
)
fig.show()
```

```python
# üßÆ Classify Inventory Health
def classify_inventory(row):
    if row['EndInventory'] < 5:
        return 'Low Stock'
    elif row['EndInventory'] > 100:
        return 'Overstock'
    else:
        return 'Healthy'

df['InventoryHealth'] = df.apply(classify_inventory, axis=1)

# üîç Summary
health_counts = df['InventoryHealth'].value_counts()
print("üßæ Inventory Health Summary:")
display(health_counts)

# üìä Visualize Health Distribution
import plotly.express as px

fig = px.histogram(df, x='InventoryHealth', color='InventoryHealth',
                   title='üì¶ Inventory Health Distribution',
                   labels={'InventoryHealth': 'Category'}, nbins=3)
fig.show()
```

```python
import numpy as np

# Prepare dataframe
price_demand_df = sales_df[['InventoryId', 'SalesQuantity', 'SalesPrice']].copy()

# Clean data: remove zero or negative prices/quantities to avoid log errors
price_demand_df = price_demand_df[(price_demand_df['SalesPrice'] > 0) & (price_demand_df['SalesQuantity'] > 0)]

# Add log columns
price_demand_df['log_qty'] = np.log(price_demand_df['SalesQuantity'])
price_demand_df['log_price'] = np.log(price_demand_df['SalesPrice'])

# Group by InventoryId and calculate price elasticity using correlation of logs
elasticity_df = price_demand_df.groupby('InventoryId').apply(
    lambda g: pd.Series({
        'Elasticity': g['log_qty'].corr(g['log_price']),
        'AvgPrice': g['SalesPrice'].mean(),
        'AvgQty': g['SalesQuantity'].mean()
    })
).reset_index()

# Output sample
print("üìä Sample elasticity data:")
display(elasticity_df.head(10))
```

```python
# Drop rows where elasticity could not be computed (NaN)
valid_elasticity_df = elasticity_df.dropna()

# Optional: filter further to only products with absolute elasticity > 0.1
valid_elasticity_df = valid_elasticity_df[valid_elasticity_df['Elasticity'].abs() > 0.1]

# Sort by highest elasticity (most negative = most price-sensitive)
valid_elasticity_df = valid_elasticity_df.sort_values(by='Elasticity')

print("üìâ Top price-sensitive products:")
display(valid_elasticity_df.head(10))

print("üìà Least price-sensitive products:")
display(valid_elasticity_df.tail(10))
```

```python
def price_suggestion(row):
    elasticity = row['Elasticity']
    price = row['AvgPrice']
    
    if elasticity < -1.0:
        return price * 0.90  # 10% reduction
    elif -1.0 <= elasticity < -0.5:
        return price * 0.95  # 5% reduction
    elif -0.5 <= elasticity <= 0.5:
        return price         # no change
    else:
        return price * 1.05  # 5% increase

# Add SuggestedPrice column
valid_elasticity_df['SuggestedPrice'] = valid_elasticity_df.apply(price_suggestion, axis=1)

# Show samples
print("üí° Sample Price Suggestions:")
display(valid_elasticity_df[['InventoryId', 'Elasticity', 'AvgPrice', 'SuggestedPrice']].head(10))
```

```python
def determine_alert_level(row):
    if row['RecommendedReorder'] > 10 and row['Elasticity'] < -0.5:
        return 'üî• High Priority'
    elif row['RecommendedReorder'] > 5:
        return '‚ö†Ô∏è Medium Priority'
    else:
        return '‚úÖ Low Priority'

# Merge enriched_inventory data with elasticity & price suggestion
merged_df = df.merge(valid_elasticity_df[['InventoryId', 'Elasticity', 'AvgPrice', 'SuggestedPrice']], 
                     on='InventoryId', how='left')

# Create AlertLevel
merged_df['AlertLevel'] = merged_df.apply(determine_alert_level, axis=1)

# Show top 20 alerts
alerts = merged_df.sort_values(by=['AlertLevel', 'RecommendedReorder'], ascending=[True, False]).head(20)

print("üö® Priority Restock & Pricing Alerts:")
display(alerts[['InventoryId', 'RecommendedReorder', 'Elasticity', 
                'AvgPrice', 'SuggestedPrice', 'AlertLevel']])
```

```python
import pandas as pd

# Replace with your actual path if needed
df = pd.read_csv(r'C:\Users\user\Desktop\wine_inventory\enriched_inventory.csv')

# Preview to confirm
print(df.columns)
display(df.head())
```

```python
import numpy as np
from datetime import timedelta

# Create a copy for trend visualization
df_trend = df.copy()

# Create synthetic monthly dates (6 months)
months = pd.date_range(end=pd.Timestamp.today(), periods=6, freq='M')

# Repeat rows for each month
df_trend = df_trend.loc[df_trend.index.repeat(len(months))].reset_index(drop=True)

# Add SalesDate (cycled monthly dates)
df_trend['SalesDate'] = months.tolist() * len(df)

# Add synthetic NetDemand variation
np.random.seed(42)
df_trend['NetDemand'] = df_trend['NetDemand'] + np.random.randint(-10, 10, size=len(df_trend))

# Save to trend_df
trend_df = df_trend.copy()

print("‚úÖ trend_df created with SalesDate column.")
display(trend_df.head())
```

```python
import ipywidgets as widgets
from IPython.display import display
import plotly.graph_objects as go

# Dropdown widgets
inv_dropdown = widgets.Dropdown(
    options=trend_df['InventoryId'].unique(),
    description='Inventory ID:',
    layout=widgets.Layout(width='50%')
)

store_dropdown = widgets.Dropdown(
    options=trend_df['Store'].unique(),
    description='Store:',
    layout=widgets.Layout(width='50%')
)

brand_dropdown = widgets.Dropdown(
    options=trend_df['Brand'].unique(),
    description='Brand:',
    layout=widgets.Layout(width='50%')
)

# Plotting function
def plot_trend(inv_id, store, brand):
    subset = trend_df[
        (trend_df['InventoryId'] == inv_id) &
        (trend_df['Store'] == store) &
        (trend_df['Brand'] == brand)
    ]

    if subset.empty:
        print("‚ö†Ô∏è No data found for selected combination.")
        return

    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=subset['SalesDate'],
        y=subset['NetDemand'],
        mode='lines+markers',
        name='Net Demand'
    ))

    fig.update_layout(
        title=f'üìà Net Demand Over Time: {inv_id}',
        xaxis_title='Date',
        yaxis_title='Net Demand',
        height=450
    )

    fig.show()

# Interactive widget
widgets.interactive_output(
    plot_trend,
    {
        'inv_id': inv_dropdown,
        'store': store_dropdown,
        'brand': brand_dropdown
    }
)

# Display the dropdowns
display(inv_dropdown, store_dropdown, brand_dropdown)
```

```python
def classify_alert(row):
    if row['NetDemand'] < -50 and row['EndInventory'] < 20:
        return 'Reorder Urgently'
    elif row['NetDemand'] < -20 or row['EndInventory'] < 40:
        return 'Monitor Closely'
    else:
        return 'Healthy'

df['InventoryAlert'] = df.apply(classify_alert, axis=1)

# üìã See some sample alerts
print("üì¢ Sample Inventory Alerts:")
display(df[['InventoryId', 'Store', 'EndInventory', 'NetDemand', 'InventoryAlert']].head(10))
```

```python
import plotly.express as px
from ipywidgets import interact

# Unique list of stores for dropdown
store_options = df['Store'].unique()

# Interactive function to show alert distribution per store
@interact
def alert_summary(store=store_options):
    filtered = df[df['Store'] == store]
    alert_counts = filtered['InventoryAlert'].value_counts().reset_index()
    alert_counts.columns = ['AlertLevel', 'Count']

    fig = px.bar(
        alert_counts,
        x='AlertLevel',
        y='Count',
        color='AlertLevel',
        color_discrete_map={
            'Healthy': 'green',
            'Monitor Closely': 'orange',
            'Reorder Urgently': 'red'
        },
        title=f'üîç Inventory Alert Summary for Store {store}',
        height=400
    )
    fig.update_layout(xaxis_title='Alert Level', yaxis_title='Number of Items')
    fig.show()
```
